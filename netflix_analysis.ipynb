```python
import pandas as pd

df = pd.read_csv(r'C:\Users\bavis\Downloads\netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna('Unknown', inplace=True)
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar', color=['red', 'black'])
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Top 10 Countries on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Shows')
plt.xticks(rotation=45)
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='barh', color='darkred')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', color='red')
plt.title('Netflix Content Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Shows')
plt.show()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar', color='darkred')
plt.title('Top 10 Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
